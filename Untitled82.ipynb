Wiener Filter for Image Restoration:
The Wiener filter is a method for image restoration that can be used to recover an image from its noisy version. You can implement the Wiener filter using libraries like SciPy in Python. You'll need the original image, the noisy image, and the estimated power spectral density (PSD) of the noise.

Here are the methods of how to apply the Wiener filter:

```python
import numpy as np
from scipy.signal import convolve2d
from scipy.signal import fftconvolve

def wiener_filter(original, noisy, noise_psd):
    H = fftconvolve(original, np.rot90(original, 2), 'same')
    return np.real(fftconvolve(noisy, np.rot90(original, 2), 'same') / (H + noise_psd))
```

```python
urls = ["/content/IMG_20230927_030732.jpg"]
for url in urls:
  image = cv2.imread(url)
```

```python
#loading the images and ads the noise
noisy_image = image + np.random.normal(0, 25, image.shape).astype(np.uint8)
```

```python
# Estimate the PSD of the noise (you may need to adapt this)
noise_psd = np.mean(np.abs(np.fft.fft2(noisy_image))**2)
```

```python
# Apply the Wiener filter
restored_image = wiener_filter(image, noisy_image, noise_psd)
```

```python

# Display the noisy and restored images
cv2_imshow(noisy_image)
cv2_imshow(restored_image)


```
